```python
import numpy as np
import pandas as pd
import jellyfish
```

```python
fifa_squads = pd.read_csv("../input/2018-fifa-world-cup-squads/2018 FIFA World Cup Squads.csv")[['Player', 'Club']]
fifa18 = pd.read_csv('../input/ea-fifa18/compressed.csv')
```

```python
fifa_squads.head()
```

```python
fifa_squads['Player'] = fifa_squads['Player'].apply(lambda x:x.replace(' (captain)',''))
```

```python
fifa18.head()
```

```python
import time
ids = []
 
start_time = time.time()
for index, row in fifa_squads.iterrows():
    
    distances = []
    for index2, row2 in fifa18.iterrows():
        #dist = jellyfish.jaro_winkler("{}{}".format(row['Player'], row['Club']),"{}{}".format(row2['Name'], row2['Club']))
        dist = jellyfish.levenshtein_distance("{}{}".format(row['Player'], row['Club']),"{}{}".format(row2['Name'], row2['Club']))
        distances.append((row2['id'], dist, row['Player'], row2['Name']))
    distances = sorted(distances, key=lambda x: x[1], reverse=False)
    id, dist,  name, name2 = distances[0]
    ids.append({"Name":name, "Linked Name":name2, "id":id, 'dist':dist})
    #print("{}->{}:{} Distance:{} Will take {}s".format(name, name2, id, dist, time_*732))

time_ = time.time() - start_time 
print("Took {}s".format(time_))

ids_df = pd.DataFrame(ids)
```
